```python
from confluent_kafka import Producer
import json
```

```python
from config.kafka_secrets import KAFKA_BOOTSTRAP_SERVERS, KAFKA_API_KEY, KAFKA_API_SECRET

# print(KAFKA_BOOTSTRAP_SERVERS)
# print(KAFKA_API_KEY)
# print(KAFKA_API_SECRET)
api_key = KAFKA_API_KEY

api_secret = KAFKA_API_SECRET

bootstrap_server = KAFKA_BOOTSTRAP_SERVERS
```

```python
from confluent_kafka import Producer
import json

conf = {
    'bootstrap.servers': bootstrap_server,
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': api_key,
    'sasl.password': api_secret
}

producer = Producer(conf)

topic = "ecommerce_events_v2"

for i in range(100000):
    data = {
        "order_id": i,
        "user_id": f"user_{i}",
        "amount": 110 + i
    }

    producer.produce(
        topic=topic,
        key=str(data["order_id"]),   # ✅ THIS IS THE FIX
        value=json.dumps(data)
    )

# ✅ IMPORTANT LINE
producer.flush()

print("Data sent successfully ✅")
```
